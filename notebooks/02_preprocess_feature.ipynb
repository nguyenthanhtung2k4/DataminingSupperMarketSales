# 02 Preprocess And Feature Engineering

Notebook nay lam sach du lieu va tao cac bang feature trung gian.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.config import load_params, resolve_path
from src.data import load_raw_data, clean_sales_data
from src.features import build_basket_transactions, build_customer_features, build_weekly_sales

config = load_params()
raw_df = load_raw_data(config, PROJECT_ROOT)
clean_df, cleaning_report = clean_sales_data(raw_df, config)
transactions_df = build_basket_transactions(clean_df, item_level=config['preprocessing']['basket_item_level'])
customer_features = build_customer_features(clean_df)
weekly_sales = build_weekly_sales(clean_df, frequency=config['forecasting']['frequency'])

processed_dir = resolve_path(config['paths']['processed_dir'], PROJECT_ROOT)
processed_dir.mkdir(parents=True, exist_ok=True)
clean_df.to_csv(processed_dir / 'cleaned_sales.csv', index=False)
transactions_df.to_csv(processed_dir / 'transactions.csv', index=False)
customer_features.to_csv(processed_dir / 'customer_features.csv', index=False)
weekly_sales.to_csv(processed_dir / 'weekly_sales.csv', index=False)
print(cleaning_report)
```
